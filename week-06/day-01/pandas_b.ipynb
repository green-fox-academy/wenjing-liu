```python
#Find apps, that is in the top 25% of apps based on ratings and were updated in the last two months of dataset.
import datetime
num = np.ceil(play_store_data['Rating'].count()*0.25).astype(int)
start = play_store_data.sort_values(by=['Last Updated'], ascending=False).iloc[0]['Last Updated']+np.timedelta64(1,'D')
end = start - np.timedelta64(60,'D')

apps_25 = play_store_data.sort_values(by=['Rating', 'Last Updated'], ascending=False)[0:num]

result = apps_25[(apps_25['Last Updated'] < pd.Timestamp(start)) &
                (apps_25['Last Updated'] > pd.Timestamp(end))]
result
```

```python
# Which app not to download? 
# - Find an app which takes the most space on your phone, but is rated poorly (define poorly)?
avg_rating = play_store_data['Rating'].mean()

def poorly(rating):
    return np.where(rating < avg_rating * 0.25,'Yes','No')

def convert_size_unit(x):
    if x[-1] == 'M':
        return float(x[:-1])*1000
    elif x[-1] == 'k':
        return float(x[:-1])
    elif x == 'Varies with device':
        return 0
    else:
        return x

play_store_data['Size In K'] = play_store_data['Size'].apply(convert_size_unit)

poor_apps = play_store_data[play_store_data['Rating']<avg_rating * 0.25].sort_values(by=['Size In K'], ascending=False)
result = poor_apps[:1]
result
```

```python
# Create a reviews reliability of reviews index, which takes into account, 
# how many reviews does the app have and how high the rating is. What metric do you take? Create as a new field.
avg_rating = play_store_data['Rating'].mean()
bottom_rating = avg_rating * 0.25
top_rating = 5
bottom_review_num = 50

def is_reliable(rating, review_count):
    if rating > bottom_rating and rating < top_rating and review_count > bottom_review_num:
        return True
    else:
        return False

play_store_data['Is Reliable Rating Reviews'] = play_store_data.apply(lambda row: 
                                                                      is_reliable(row['Rating'], row['Reviews']),
                                                                     axis=1)
play_store_data
```
